```python
raw_data=open('/content/drive/MyDrive/Rubixe/TF-IDF/1661-0.txt', encoding='utf-8').read().lower()
```

```python
raw_data=raw_data.translate(str.maketrans("", "", string.punctuation))
```

```python
data=pd.DataFrame({"previous_words":[], "next_word":[]})
```

```python
list_data=raw_data.split()
print(len(list_data))
```

```python
group_list=list()
```

```python
for i in range(len(list_data)-3):
    group_list.append([" ".join(list_data[i:i+3]), list_data[i+3]])
```

```python
print(group_list[:5])
```

```python
print(group_list[0][0])
```

```python
for i in range(len(group_list)):
    data.loc[len(data)]={'previous_words':group_list[i][0], 'next_word':group_list[i][1]}
```

```python
data.head()
```

# TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
uni=dict()
for i, word in enumerate(np.unique(data.next_word)):
    uni[word]=i
```

```python
len(uni)
```

```python
data['next_word_num']=data['next_word'].map(uni)
```

```python

data.head()
```

```python
x_train, y_train=data.previous_words, data.next_word_num 
```

```python
x_train.shape, y_train.shape
```

```python
from sklearn.pipeline import Pipeline
from sklearn.svm import SVC
from sklearn.metrics import classification_report
```

```python
clf=Pipeline([
    ('vectorizer_tfidf', TfidfVectorizer()),
    ("svm", SVC(kernel='poly'))
])

clf.fit(x_train, y_train)
```

```python
y_pred = clf.predict(x_train)
```

```python
print(clf.score(x_train, y_train))
```

```python
print(y_pred[:5], y_train[:5])
```
